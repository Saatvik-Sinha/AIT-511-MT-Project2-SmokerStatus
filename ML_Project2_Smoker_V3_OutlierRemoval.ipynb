```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train_dataset.csv')
df
```

# Preprocessing

```python
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
df = df[~((df['hearing(left)']==1) & (df['smoking']==0) & (df['age']>75))]
df = df[~((df['hearing(left)']==1) & (df['smoking']==1) & (df['age']>70))]
df = df[~((df['hearing(left)']==2) & (df['smoking']==0) & (df['age']<35))]
df = df[~((df['hearing(left)']==2) & (df['smoking']==1) & (df['age']<30))]

df = df[~((df['hearing(left)']==1) & (df['smoking']==0) & (df['height(cm)']<135))]
df = df[~((df['hearing(left)']==1) & (df['smoking']==1) & (df['height(cm)']<150))]
df = df[~((df['hearing(left)']==2) & (df['smoking']==1) & (df['height(cm)']<145))]

df = df[~((df['hearing(left)']==2) & (df['smoking']==1) & (df['weight(kg)']>100))]

df = df[~((df['hearing(left)']==2) & (df['smoking']==0) & (df['Cholesterol']>300))]

df = df[~((df['hearing(left)']==1) & (df['triglyceride']>400))]

df = df[~((df['hearing(left)']==1) & (df['smoking']==0) & (df['HDL']>350))]

df = df[~((df['hearing(left)']==2) & (df['hemoglobin']<11))]

df = df[~((df['hearing(left)']==2) & (df['smoking']==0) & (df['ALT']>2500))]
```

```python
df = df[~((df['Urine protein']==5) & (df['smoking']==1) & (df['age']>65))]

df = df[~((df['Urine protein']==1) & (df['smoking']==0) & (df['height(cm)']<135))]
df = df[~((df['Urine protein']==1) & (df['smoking']==1) & (df['height(cm)']<150))]
df = df[~((df['Urine protein']==3) & (df['smoking']==1) & (df['height(cm)']<150))]

df = df[~((df['Urine protein']==5) & (df['smoking']==1) & (df['relaxation']>130))]

df = df[~((df['Urine protein']==5) & (df['smoking']==1) & (df['fasting blood sugar']>150))]
```

```python
df
```

```python
x = df.drop(columns=['smoking'])
y = df['smoking']
```

```python
x[['hearing(left)', 'hearing(right)']] = x[['hearing(left)','hearing(right)']] - 1

x = pd.get_dummies(x, columns=['Urine protein'], prefix='urine_protein')
```

```python
x
```

# Models

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
```

```python
categorical_cols = ['hearing(left)', 'hearing(right)', 'Urine protein', 'dental caries', 'smoking']
continuous_cols = [col for col in df.columns if col not in categorical_cols]

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
pp = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), continuous_cols)
    ],
    remainder='passthrough'
)
```

## Logistic Regression

```python
lrpipe = Pipeline([
    ('preprocess', pp),
    ('clf', LogisticRegression())
])

lrparam = {
    'clf__penalty': [None, 'elasticnet', 'l1', 'l2'],
    'clf__C': [0.1, 0.3, 1, 3, 10, 30],
    'clf__l1_ratio': [0.25, 0.5, 0.75],
    'clf__solver': ['saga', 'lbfgs']
}
```

```python
gridlr = GridSearchCV(estimator=lrpipe, param_grid=lrparam, scoring='accuracy', cv=skf, verbose=3, return_train_score=True)
gridlr.fit(x, y)
```

```python
print(gridlr.best_score_)
print(gridlr.best_params_)

bestlr = gridlr.best_estimator_
ypred = bestlr.predict(x)
print(accuracy_score(y, ypred))
print(classification_report(y, ypred))
print(confusion_matrix(y, ypred))
```

## Support Vector Classifier

```python
svcpipe = Pipeline([
    ('preprocess', pp),
    ('clf', SVC(max_iter=100000, kernel='rbf'))
])

svcparam = {
    'clf__C': [0.1, 0.3, 1, 3, 10],
}
```

```python
gridsvc = GridSearchCV(estimator=svcpipe, param_grid=svcparam, scoring='accuracy', cv=skf, verbose=3, return_train_score=True)
gridsvc.fit(x, y)
```

```python
print(gridsvc.best_score_)
print(gridsvc.best_params_)

bestsvc = gridsvc.best_estimator_
ypred = bestsvc.predict(x)
print(accuracy_score(y, ypred))
print(classification_report(y, ypred))
print(confusion_matrix(y, ypred))
```

## Dense Neural Network

```python
nnskf = StratifiedKFold(n_splits=3, shuffle=False)

nnpipe = Pipeline([
    ('preprocess', pp),
    ('clf', MLPClassifier(solver='adam', batch_size=8192, tol=0.1, early_stopping=True, random_state=42, validation_fraction=0.25))
])

nnparam = {
    'clf__learning_rate_init': [0.001, 0.003, 0.01, 0.03, 0.1, 0.3],
    'clf__alpha': [0, 0.003, 0.01, 0.03, 0.1, 0.3, 1, 3],
    'clf__hidden_layer_sizes': [(64,), (128,), (64, 64), (128, 128)],
    'clf__activation': ['logistic', 'tanh', 'relu'],
}
```

```python
gridnn = GridSearchCV(estimator=nnpipe, param_grid=nnparam, cv=nnskf)
gridnn.fit(x, y)
```

```python
print(gridnn.best_score_)
print(gridnn.best_params_)

bestnn = gridnn.best_estimator_
ypred = bestnn.predict(x)
print(accuracy_score(y, ypred))
print(classification_report(y, ypred))
print(confusion_matrix(y, ypred))
```
